```python
business_data = pd.read_csv('../dataset/business.csv', dtype=unicode, usecols=["business_id" ,"name" ,"city", "categories"])
review_data = pd.read_csv('../dataset/review.csv', usecols=['review_id', 'business_id', 'stars', 'text'])
```

```python
review_data[0:3]
```

```python
list(business_data)
```

```python
business_review_data = business_data.merge(review_data, left_on='business_id', right_on='business_id', how='outer')

```

```python
list(business_review_data)
```

```python
restaurants_business_data = business_review_data[business_review_data['categories'].str.contains("'Restaurants'")]
```

```python
cities = ['Las Vegas', 'Charlotte', 'Phoenix', 'Pittsburgh']
```

```python
for city in cities:
    city_restaurants_business_data = restaurants_business_data.loc[restaurants_business_data['city'] == city]
    city_restaurants_business_data = city_restaurants_business_data[['text','stars']]
    city_restaurants_business_data.to_csv(city + "_review_data.csv")
```

```python
city_restaurants_business_data
```

```python
business_group = restaurants_business_data.groupby(['city'])['review_id'].count().reset_index()
```

```python
business_group_sorted = business_group.sort_values('review_id', ascending=False)
```

```python
business_group_sorted
```

```python
business_group_sorted.to_csv("restaurant_city_reviewcount_from_reviewid.csv")
```
